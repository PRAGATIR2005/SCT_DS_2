```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['TASK 2Customer_Data.csv']))
import os
os.makedirs("charts", exist_ok=True)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')
plt.title("Gender Distribution")
plt.savefig("charts/gender_distribution.png")
plt.show()
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True, bins=10)
plt.title("Age Distribution")
plt.savefig("charts/age_distribution.png")
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(data=df, y='Annual Income (k$)')
plt.title("Annual Income Distribution")
plt.savefig("charts/income_boxplot.png")
plt.show()
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Age', y='Spending Score', hue='Gender')
plt.title("Age vs Spending Score")
plt.savefig("charts/age_vs_spending.png")
plt.show()
plt.figure(figsize=(6,4))
sns.heatmap(df.drop('CustomerID', axis=1).select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("charts/correlation_heatmap.png")
plt.show()
import shutil
shutil.make_archive("EDA_Charts", 'zip', "charts")
files.download("EDA_Charts.zip")
```
